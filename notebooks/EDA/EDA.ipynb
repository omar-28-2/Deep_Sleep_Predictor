```python
# Drop non-numeric or convert them before correlation
df_corr = mergin_cleaned_df.copy()

# Convert Gender to numeric if it's not already (e.g., Male:1, Female:0)
if df_corr['Gender'].dtype == 'object':
    df_corr['Gender'] = df_corr['Gender'].map({'Male': 1, 'Female': 0})

# Drop Occupation or one-hot encode it (correlation only works on numeric)
df_corr = df_corr.drop('Occupation', axis=1)

# Now create the correlation matrix
correlation_matrix = df_corr.corr(numeric_only=True)

# Plot heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Correlation Heatmap")
plt.show()
```
