```python
import pandas as pd
from ctgan import CTGAN

df = pd.read_csv("merge.csv")

ctgan = CTGAN()

ctgan.fit(df)

synthetic_data = ctgan.sample(1000)

synthetic_data.to_csv("synthetic_merge_1000.csv", index=False)

print(synthetic_data.head())
```

```python
df_augumented=synthetic_data.copy()
```

```python
df.describe()
```

```python
df_augumented.describe()
```

```python
df_augumented = df_augumented[(df_augumented["Physical Activity Level"] >= 0)]
```

```python
df_augumented = df_augumented[(df_augumented["Quality of Sleep"] <= 10) & (df_augumented["Sleep Duration"] <= 12) & (df_augumented["Stress Level"] <= 13) & (df_augumented["Blood Pressure"] <= 2.25) & (df_augumented["Heart Rate"] <= 105)]
```

```python
df_augumented = df_augumented[(df_augumented["Quality of Sleep"] >= 1) & (df_augumented["Sleep Duration"] >= 2) & (df_augumented["Stress Level"] >= 0) & (df_augumented["Blood Pressure"] >= 1.1) & (df_augumented["Heart Rate"] >= 44)]
```

```python
df_augumented.describe()
```

```python
df_augumented.count()
```

```python
df_augumented = pd.concat([df, df_augumented], ignore_index=True)
df_augumented
```

```python
df_augumented = df_augumented[df_augumented["Stress Level"] >= 0]
```

```python
import matplotlib.pyplot as plt

features_to_plot = ['Sleep Duration', 'Quality of Sleep','Heart Rate','Stress Level','Blood Pressure']  

for feature in features_to_plot:
    plt.figure(figsize=(6, 3))
    plt.hist(df[feature], alpha=0.5, label='Original', bins=30)
    plt.hist(df_augumented[feature], alpha=0.5, label='Synthetic', bins=30)
    plt.title(f'Distribution of {feature}')
    plt.legend()
    plt.show()
```

```python
df_augumented.to_csv("augmented_merge_CTGAN.csv", index=False)
```
